```python
#Importing Libraries
%config IPCompleter.greedy=True
# Common imports
import numpy as np
import os
import pandas as pd
from sklearn import metrics
import matplotlib.pyplot as plt
import logomaker
```

```python
#Read data
peptide_only = pd.read_csv("Visual_Peptides - Copy.csv")

#Convert to series
Peptideseries=peptide_only.iloc[0:100,0]
type(Peptideseries)

PeptideseriesConCatenate=Peptideseries.str.cat(sep='').replace('\t','')
PeptideseriesConCatenate

counts_aa = logomaker.alignment_to_matrix(Peptideseries)
counts_aa.head()
```

```python
#Logo with defalt settings
logomaker.Logo(counts_aa)
```

```python
pept_logo = logomaker.Logo(counts_aa,
                          color_scheme='chemistry',
                          shade_below=.5,
                          fade_below=.5,
                          font_name='Arial Rounded MT Bold')

# style using Logo methods
pept_logo.style_spines(visible=False)
pept_logo.style_spines(spines=['left', 'bottom'], visible=True)
pept_logo.style_xticks(rotation=90, fmt='%d', anchor=0)

# style using Axes methods
pept_logo.ax.set_ylabel("Count", labelpad=-1)
pept_logo.ax.set_xlabel("Amino Acids", labelpad=-1)
pept_logo.ax.xaxis.set_ticks_position('none')
pept_logo.ax.xaxis.set_tick_params(pad=-1)

# style and show figure
pept_logo.fig.show()
```
